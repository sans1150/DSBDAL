```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer

# Download required packages
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')

# Sample document
text = "Data Analytics is an important part of Data Science."

print("Original Text:")
print(text)

# Tokenization
tokens = word_tokenize(text)

print("\nTokens:")
print(tokens)

# POS Tagging
pos = pos_tag(tokens)

print("\nPOS Tagging:")
print(pos)

# Stopwords Removal
stop_words = stopwords.words('english')

filtered = [word for word in tokens
            if word.lower() not in stop_words]

print("\nAfter Stopwords Removal:")
print(filtered)

# Stemming
ps = PorterStemmer()

stemmed = [ps.stem(word) for word in filtered]

print("\nStemmed Words:")
print(stemmed)

# Lemmatization
lm = WordNetLemmatizer()

lemmatized = [lm.lemmatize(word) for word in filtered]

print("\nLemmatized Words:")
print(lemmatized)

# TF-IDF Representation

documents = [
    "Data Analytics is important",
    "Data Science uses Analytics",
    "Machine Learning is part of Data Science"
]

tfidf = TfidfVectorizer()

result = tfidf.fit_transform(documents)

print("\nTF-IDF Matrix:")
print(result.toarray())

print("\nFeature Names:")
print(tfidf.get_feature_names_out())
```
