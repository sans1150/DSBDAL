```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Display first rows
print(df.head())

# Dataset information
print(df.info())

# Histogram of Fare
plt.figure(figsize=(8,5))

sns.histplot(df['fare'], bins=30)

plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Count")

plt.show()
```
