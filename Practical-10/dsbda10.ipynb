```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\SANSKRUTI\OneDrive\Desktop\THIRD YEAR (TE)\DSBDA CODES\data6.csv")

# Display first rows
print(df.head())

# Features and data types
print("\nFeatures and Data Types")

print(df.dtypes)

# Histograms for all features

df.hist(figsize=(10,8))

plt.show()

# Boxplots for all numeric features

plt.figure(figsize=(10,6))

sns.boxplot(data=df.iloc[:,1:5])

plt.title("Boxplot of Iris Features")

plt.show()

# ---------------------------------------------------
# Compare distributions and identify outliers
# ---------------------------------------------------

print("\nOutlier Detection")

for col in df.iloc[:,1:5].columns:

    Q1 = df[col].quantile(0.25)

    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR

    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ][col]

    print(f"\n{col}")

    print("Outliers :", outliers.values)
```
